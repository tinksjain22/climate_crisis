```python
from pathlib import Path
import pandas as pd
import scipy.stats as stats
import csv
import matplotlib.pyplot as plt
```

```python
# Create a file path
global_temp = Path("all countries global temperature.csv")
CO2_emission = Path("historical_emissions.csv")
population = Path("population_data.csv")

# Read in the data. 
temp_df = pd.read_csv(global_temp)
CO2_df = pd.read_csv(CO2_emission)
pop_df = pd.read_csv(population)

temp_df.head()
```

```python
#cleaning csv's
temp_df.drop(['ObjectId', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980',
              '1981', '1982', '1983', '1984', '1985','1986', '1987', '1988', '1989'], axis=1, inplace=True)
CO2_df.drop(['Data source', 'Sector'], axis=1, inplace=True)
pop_df.drop(['Country Code', '1960','1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969',
             '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980',
             '1981', '1982', '1983', '1984', '1985','1986', '1987', '1988', '1989'], axis=1, inplace=True)
```

```python
CO2_df.drop(['Gas', 'Unit'], axis=1, inplace=True)
CO2_df.head()
```

```python
# Graph attempt; need to melt dataframe for plotting
CO2_df_melted = CO2_df.cmelt(id_vars='Country', var_name='Year', value_name='CO2 Emissions')

# Convert year column to numeric
CO2_df_melted['Year'] = pd.to_numeric(CO2_df_melted['Year'])

# Create the time series plot
CO2_df_filtered = CO2_df_melted[CO2_df_melted['Country'] != 'World']
plt.figure(figsize=(12, 6))
for country in CO2_df_filtered['Country'].unique():
    country_data = CO2_df_filtered[CO2_df_filtered['Country'] == country]
    plt.plot(country_data['Year'], country_data['CO2 Emissions'], label=country)

plt.title('CO2 Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions')
plt.show()
```
